```python
key = "CURRFX/USDCNY"
```

```python
import quandl
```

## Get the last 6 months' data

```python
values = quandl.get(key, returns="numpy", rows=30*6, order="desc")
```

```python
values[0]
```

```python
def plot(data, n_samples=30):
    xs = [x[0] for x in data[:n_samples]]
    ys = [x[1] for x in data[:n_samples]]
    plt.plot(xs, ys)
```

## Last week

```python
plot(values, 7)
```

## Last 30 days

```python
plot(values, 30)
```

## Last 3 months

```python
plot(values, 30*3)
```

## Last 6 months

```python
plot(values, 30*6)
```
